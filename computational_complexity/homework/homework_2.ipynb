Задача 3-4 г
------------

Будем доказывать, что задача возведения квадратной матрицы в степень k лежит в классе $AC^1$

Воспользуемся фактом, что задача сложения 2-х чисел и умножения 2-х матриц лежит в $AC^0$

Объяснение этого факта например можно найти на $49$ странице [здесь](http://www.cs.umd.edu/~jkatz/complexity/f11/all.pdf)

Если совсем коротко, то $(AB)_{i, j} = \bigvee_{1 \leq k \leq N} (A_{i, k} \wedge B_{k, j})$

Еще нам понадобится деление числа, записанного в двоичной записи на $2$ максимально близких числа. То есть, если $k$ - четное, то это $k/2$ и $k/2$. Если $k$ - нечетное, то это $k/2$ и $k/2+1$, где $/$ - это целочисленное деление. Опишем этот алгоритм. Если младший бит числа $0$, то число четное, и деление на $2$ эквивалентно стиранию младшего бита. Если последний бит равен $1$, то число нечетное, число $k/2$, мы получим стиранием младшего бита, а число $k/2 + 1$ мы вычислим прибавлением $1$. Сложение любых 2-х чисел лежит в $AC^0$, а все наши рассуждения выражаются конечным набором логических связок. Значит деление на $2$ числа лежит в $AC^0$.

Теперь вспомним алгоритм бинарного возведения в степень.

```
POW(A, k):
    Если k == 0:
        return E    
    Если k - четное:
        B = POW(A, k/2)
        return B * B
    иначе:
        return POW(A, k-1) * A    
```

Теперь соберем нашу схему по этому алгоритму. Важно понимать, что алгоритм выше реккурсивный. А наша схема его последовательная реализация, то есть схема будет работать начиная с маленьких степеней $A$, переходя к большим. Всего нужно не более $2 * log(k)$ шагов и на каждом шаге, мы используем схему $AC^0$. Значит вся схема лежит в $AC^1$

Задача 5е
----------

Если слово принадлежит языку $P(\text{правильного ответа}) > \alpha$, а $P(\text{неправильного ответа}) \leq \alpha$

Если слово не принадлежит $P(\text{неправильного ответа}) \leq \alpha$, a $P(\text{правильного ответа}) > \alpha$

С учетом полиномиальности формулировка похожа на PPTIME (в оригинале было нестрогое неравенство)

Вспомним рассуждение с семинара про [неравенство Чернова](http://en.wikipedia.org/wiki/Chernoff_bound) о том, что $\alpha$ можно заменить на любое число из $1/2 + \delta \in (0, 1) : \delta \geq 0$

Обозначим за $X_i$ - случайную величину, равную правильному ответу машины на слове $x$ на $i-ом$ запуске. Они независимы и одинаково распределены. 
Запустим машину $n$ раз, и выдадим ответ, который наиболее часто выдавала машина. Обозначим за $p = 1/2 - \delta$

$$
\text{Тогда }P(\text{неправильного ответа}) = P(\sum_{i=1}^{n} X_i \leq n/2) = P(\sum_{i=1}^{n} (1-X_i) \geq n/2) = P(S_{n} \geq pn(\frac{1/2}{1/2-\delta})) \leq e^{-\frac{\delta^2 * n}{3(1/2-\delta)}}
$$

$$
\text{Например, чтобы ошибка не превысила} \frac{1}{3} \text{нужно}
$$

$$
n \geq \frac{3*ln(3)*(1/2-\delta)}{\delta^2} \text{ и новое } N' = \frac{3*ln(3)}{2*\delta^2}
$$  

Количество запусков не зависит от длины слова и время работы увеличивается в константу, а значит до сих пор укладывается в полином.

Получается эквивалентными можно считать все классы при $\alpha \geq 1/2$ и $\alpha < 1/2$ (доказывается по аналогии с заменой на противоположный ответ). А вот между собой эти 2 класса не эквивалентны, так как, чтобы нам получить эквивалентность $PP_\alpha$ и $PP_{1-\alpha}$ нужно всегда инвентировать ответ машины из второго класса, значит ответ на задачу нет.

Задача 6е
----------

Задача поиска количества полных паросочетаний в двудольном графе размера $2*n$ решается вычислением перманента матрицы размером $n*n$

Пронумеруем вершины первой доли числами от $1$ до $n$, и вершины второй доли числами от $1$ до $n$.
Построим матрицу $A$, где элементом матрицы $a_{i, j}$ является $1$, если вершина с номером $i$ из первой доли имеет обшее ребро с вершиной $j$ из второй доли. Матрица $A$ имеет размер $n*n$. Посмотрим на её перманент.

По определению перманент:

$perm(A) = \prod_{\sigma} \sum_{i=1}^{n} a_{i, \sigma(i)}$

Каждая перестановка соответствует возможному паросочетанию, где будут соединены ребрами $i$ и $\sigma(i)$. Если все $a_{i, \sigma(i)}$ будут равны $1$, то такое паросочтание возможно, и оно войдет в общую сумму. Значит перманент матрицы A является количеством паросочетаний в двудольном графе.

Так как задача вычисления перманента $NP-трудна$, то и наша задача $NP-трудна$

Осталось заметить, что задача поиска полного паросочетания в двудольном графе - классическая задача из $P$, решаемая алгоритмом Куна, например алгоритм можно прочитать [здесь](http://neerc.ifmo.ru/wiki/index.php?title=%D0%90%D0%BB%D0%B3%D0%BE%D1%80%D0%B8%D1%82%D0%BC_%D0%9A%D1%83%D0%BD%D0%B0_%D0%B4%D0%BB%D1%8F_%D0%BF%D0%BE%D0%B8%D1%81%D0%BA%D0%B0_%D0%BC%D0%B0%D0%BA%D1%81%D0%B8%D0%BC%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B3%D0%BE_%D0%BF%D0%B0%D1%80%D0%BE%D1%81%D0%BE%D1%87%D0%B5%D1%82%D0%B0%D0%BD%D0%B8%D1%8F)
