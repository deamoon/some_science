=========================

Задача 3п
---------

Докажите NP-полноту

$$
MAX2SAT=\{(\varphi, k) | \varphi - \text{формула в виде 3-КНФ, такая что для некоторого набора значений переменных выполнено хотя бы k дизъюнктов}\}.
$$

Заметим, что эта задача принадлежит NP, так как если мы получим набор переменных, мы легко сможем проверить выполнимость k дизъюнктов за один проход, то  есть за полиномиальное время.

Сведем известную NP-полную задачу 3SAT к MAX2SAT.

Пусть на вход мы получили формулу $\varphi$, преобразуем каждый её дизъюнкт по следующему правилу:

$$
(x_1 \vee x_2 \vee x_3) \to [(x_1)\wedge(x_2)\wedge(x_3)\wedge(\omega)]\wedge[(\bar{x_1}\vee\bar{x_2})\wedge(\bar{x_2}\vee\bar{x_3})\wedge(\bar{x_1}\vee\bar{x_3})]\wedge[(x_1\vee\bar{\omega})\wedge(x_2\vee\bar{\omega})\wedge(x_3\vee\bar{\omega})]
$$

Получим $R(\varphi)$. Тогда ответом на исходную задачу будет $MAX2SAT(R(\varphi), 7 \times n)$, где $n$ - это количество дизъюнктов в формуле $\varphi$

Количество скобок увеличилось в 10 раз, поэтому эта сводимость полиномиальная

Доказательство
-------------

Проанализируем как ведет себя дизъюнкт после замены

1. Если в исходной формуле были истинны $x_1, x_2, x_3$, тогда после замены максимальное количество истинных дизъюнктов - это 7, оно достигается, если сделать $\omega$ истинной
2. Если в исходной формуле были истинны только 2 из $x_1, x_2, x_3$, тогда после замены максимальное количество истинных дизъюнктов также 7, оно достигается, если сделать $\omega$ ложной
3. Если в исходной формуле были истинны только одна из $x_1, x_2, x_3$, тогда после замены максимальное количество истинных дизъюнктов также 7, оно достигается, если сделать $\omega$ ложной
4. Если в исходной формуле ни одна из $x_1, x_2, x_3$ не было истинной, тогда после замены максимальное количество истинных дизъюнктов только 6, оно достигается, если сделать $\omega$ ложной

Значит если $MAX2SAT(R(\varphi), 7 \times n) = TRUE$, то в $R(\varphi)$ нашлось 7n истинных дизъюнктов, в силу построения, это означает, что в каждом дизъюнкте $\varphi$ нашлась хотя бы одна истинная переменная и $\varphi$ выполнима.

Если же $MAX2SAT(R(\varphi), 7 \times n) = FALSE$, то в $R(\varphi)$ не нашлось 7n истинных дизъюнктов, в силу построения, это означает, что для любого набора существует дизъюнкт в $\varphi$, такой что все 3 переменные в нем ложны и $\varphi$ невыполнима

Задача 5c
--------

Докажите NL-полноту

$$
0-1-INEQ = \{S | D - \text{ совместная система неравенств вида } x <= y \text{ и равенств вида } x = a, \text{ где все переменные и константы принимают булевы значения}\}
$$

Задача лежит в классе NL.
Сопоставим каждой переменной вершину, а отношению x <= y направленное ребро (x, y).
Теперь, чтобы проверить истинность системы достаточно проверить что в путях, начинающихся от истинных переменных, мы не можем прийти в ложную. И то что, у нас нет противоречивых равенств одной переменной. Для обхода такого графа, мы храним всегда несколько вершин. 

Теперь сведем задачу PATH(G, s, t) - {существует ли путь в ориентированном графе G из s в t} к нашей задаче.

Для этого преобразуем граф G в систему неравенств по принципу, вершины станут переменными, а ребра их отношениями, ребро (x, y), значит x <= y. Также добавим s = 1, t = 0. И тогда отрицание ответа 0-1-INEQ будет ответом на PATH.

Доказательство
-------------

Заметим, что если x=1 и x<=y, то чтобы система была истинной должно быть верно y=1.

Аналогично если y=0 и x<=y, то чтобы система была истинной x=0.

Значит все вершины, достижимые из s, должны стать истинными, а все, вершины из которых достижима t, должны быть ложными.

Получается отсутствие пути эквивалентно истинности системы неравенств.

Задача 6т
---------

$$
\text{Докажите, что } NP \ne E
$$

Для начала обратимся к [лекциям](http://ru.discrete-mathematics.org/fall2014/3/complexity/lecture-2-p-np.pdf) и выпишем известные для нас факты:

1. $E = \bigcup_{k=1}^\infty DTIME(k^n)$
2. $EXP = \bigcup_{k=1}^\infty DTIME(2^{n^k})$ 
3. $E \subset EXP$
4. $NP \subseteq EXP$

**Теорема:**
Если существует вещественное чиcло $p > 1$, такое что $DTIME(p^n) \subseteq NP$, тогда $\bigcup_{k=1}^\infty DTIME(k^n) \subsetneq NP = EXP$ или по другому $E \subset NP$

**Следствие:**
Не существует вещественного $p > 1$ такого, что $DTIME(p^n) = NP$, а значит и $\bigcup_{k=1}^\infty DTIME(k^n) \neq NP$, то есть $NP \ne E$

**Доказательство следствия:**

От противного, 
$$
DTIME(p^n) = NP \Rightarrow DTIME(p^n) \subseteq NP \Rightarrow \bigcup_{k=1}^\infty DTIME(k^n) \subsetneq NP \Rightarrow  DTIME(p^n) \subsetneq NP \Rightarrow DTIME(p^n) \neq NP
$$

**Лемма 1:**
Если существует $p > 1$ такой, что $DTIME(p^n) \subseteq NP$, тогда $DTIME((p^2)^n) \subseteq NP$

**Доказательство Леммы 1:**

Пусть $p$ существует. Возьмем $L_1 \in DTIME((p^2)^n)$, пусть $M_1$ - машина Тьюринга, распознающая этот язык за время не больше чем $(p^2)^n$. Пусть $\Sigma$ конечный алфаивт и $L_1 \subseteq \Sigma^{*}$, а $c$ - символ $c \notin \Sigma$.

Положим $L_2 = \{wc^{|w|}|w \in L_1\}$, $M_2$ машина Тьюринга, распознающая $L_2$ за время не больше, чем $(p^2)^{|w|}$. $|wc^{|w|}| = 2|w|$, значит $M_2$ работает за $p^n$. $L_2 \in DTIME(p^n) \subseteq NP$. Но тогда и $L_1 \in NP$, так как разница в распознавании $L_1$  и $L_2$ укладывается в полиномиальное время. Значит $DTIME((p^2)^n) \subseteq NP$

**Лемма 2:**
Если $\bigcup_{k=1}^\infty DTIME(k^n) \subseteq NP$, тогда $\bigcup_{k=1}^\infty DTIME(2^{n^k}) \subseteq NP$

**Доказательство Леммы 2:**

Если $L_1 \in \bigcup_{k=1}^\infty DTIME(2^{n^k})$, тогда машина Тьюринга, распознающая этот язык $M_1$ работает за $2^{n^k}$. Пусть $\Sigma$ конечный алфаивт и $L_1 \subseteq \Sigma^{*}$, а $c$ - символ $c \notin \Sigma$.

Положим $L_2 = \{wc^m|w \in L_1, |wc^m|=|w|^k\}$, $M_2$ машина Тьюринга, распознающая $L_2$ за время не больше, чем $2^{|w|^k}$. $|wc^m|=|w|^k$, значит $M_2$ работает за $2^n$.

$L_2 \in DTIME(2^n) \subseteq \bigcup_{k=1}^\infty DTIME(k^n)$. Так как $\bigcup_{k=1}^\infty DTIME(k^n) \subseteq NP$, значит $L_2 \in NP$, так как разница в распознавании $L_1$  и $L_2$ укладывается в полиномиальное время, значит $L_1 \in NP$ и  $\bigcup_{k=1}^\infty DTIME(2^{n^k}) \subseteq NP$

**Доказательство теоремы:**

Так как $E \subset EXP$, достаточно доказать, что $NP = EXP$

Пусть существует $p$ такое, что $DTIME(p^n) \subseteq NP$, используем Лемму 1, для любого $q > 1$, $DTIME((p^q)^n) \subseteq NP$. Тогда $\bigcup_{k=1}^\infty DTIME(k^n) \subseteq NP$

По Лемме 2, это значит $\bigcup_{k=1}^\infty DTIME(2^{n^k}) \subseteq NP \subseteq EXP$, тогда $NP = EXP$
